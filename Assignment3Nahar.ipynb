```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of n
n_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

# Loop through the values of n
for n in n_values:
    # Simulate throwing two dice n times
    dice_sum = np.random.randint(1, 7, size=n) + np.random.randint(1, 7, size=n)

    # Compute the frequencies of the dice sums
    hist, bins = np.histogram(dice_sum, bins=np.arange(2, 14))

    # Plot the histogram
    plt.bar(bins[:-1], hist / n)

    # Set the title of the plot
    plt.title(f"Histogram of Dice Sums (n = {n})")

    # Show the plot
    plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load data from csv file
data = pd.read_csv("weight-height.csv")

# Extract weight and height columns
weight = data["Weight"]
height = data["Height"]

plt.scatter(height, weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Weight vs Height")

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(height.values.reshape(-1, 1), weight)

# Predicted weight based on the model
predicted_weight = model.predict(height.values.reshape(-1, 1))

# Plot the actual weight vs predicted weight
plt.scatter(height, weight, label="Actual Weight")
plt.plot(height, predicted_weight, color="red", label="Predicted Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Actual vs Predicted Weight")
plt.legend()

# Calculate Root Mean Squared Error (RMSE)
rmse = mean_squared_error(weight, predicted_weight)

# Calculate R-squared coefficient of determination
r2 = r2_score(weight, predicted_weight)

# Print the results
print("RMSE:", rmse)
print("R-squared:", r2)
plt.show()
```
